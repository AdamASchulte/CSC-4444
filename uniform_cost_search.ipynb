```python
import sys
import time
def edge_cost(source, destination, cost):
    if(source == destination):
        return 0
    return cost[(source, destination)]

# uniform cost function
# returns the minimum cost (Bonus: returns path to minimum cost as well)
def  uniform_cost_search(start, goal, graph, cost):
    answer = {node: {'cost': float('inf'), 'path': []} for node in goal}

    

    #find the indices of the nodes that are actually in the graph
    non_empty_indices = [i for i, val in enumerate(graph) if val]

    #create more friendly data structure to work with, a dictionary with a key(node number) and a value(List of nodes it can reach)
    realGraph = {}

    #hydrate this dictionary with information from "graph"
    for z in range(len(non_empty_indices)):
        realGraph[non_empty_indices[z]] = graph[non_empty_indices[z]]

    #find keys of the realGraph to compare against the list of nodes I am about to find 
    allKeys = realGraph.keys()

    #add dead end nodes to my realGraph by finding every node that is touched by other nodes in graph, and adding them to the realGraph if they are not already in it.
    for iterator in range(len(graph)):
        for insideIterator in range(len(graph[iterator])):
            if(graph[iterator][insideIterator] not in allKeys):
                realGraph[graph[iterator][insideIterator]] = [-1]

    for g in goal:
        path = []

        totalFringe = {}
        #make total cost a dictionary that holds each node's number as a key and it's cost to be reached as the value
        TotalCost = {}
        currentNode = start

        #initialize all Total Costs to high value besides start
        for n in realGraph:
            TotalCost[n] = sys.maxsize
        TotalCost[start] = 0
        
        #initialize fringe with starting node's neighbors
        fringe = {}
        for x in realGraph[start]:
            fringe[x] = start
        
        while currentNode != g:
            #add new values to fringe(current node's neighbors)
            for nextNode in realGraph[currentNode]:
                if nextNode != [-1]:
                    if (nextNode != -1):
                        if(TotalCost[nextNode] > TotalCost[currentNode] + edge_cost(currentNode, nextNode, cost)):
                            TotalCost[nextNode] = TotalCost[currentNode] + edge_cost(currentNode, nextNode, cost)
                            fringe[nextNode] = currentNode

            #search for lowest cost node in fringe
            minimumAdditionalCost = sys.maxsize
            nextNode = -1
            for y in fringe:
                if TotalCost[y] < minimumAdditionalCost:
                    minimumAdditionalCost = TotalCost[y]
                    nextNode = y
            if nextNode != -1:
                totalFringe[nextNode] = fringe.pop(nextNode)
                #transition to next node
                currentNode = nextNode
            else: 
                print("Critical Error, no goal found. The program will crash in 5 seconds")
                print("Here are the previous answers plus the current failing one: " + str(answer))
                time.sleep(3)
                sys.exit()

        #set path from goal
        pathNode = g
        x = totalFringe[pathNode]
        while(pathNode != start):
            path.append(pathNode)
            print("The fringe[pathNode] is :" + str(totalFringe[pathNode]))
            pathNode = totalFringe[pathNode]
        
        path.append(start)
        path.reverse()

        #once goal is reached set current goal's cost 
        answer[g]['cost'] = TotalCost[currentNode]
        answer[g]['path'] = path

    return answer
 
# main function
if __name__ == '__main__':
     
    # create a graph with no more than 30 nodes
    graph, cost = [[] for i in range(30)], {}
 
    # add edges to the graph
    graph[0].append(4)
    graph[0].append(5)
    graph[0].append(16)
    graph[2].append(1)
    graph[3].append(1)
    graph[4].append(2)
    graph[4].append(3)
    graph[4].append(5)
    graph[5].append(8)
    graph[5].append(18)
    graph[6].append(3)
    graph[6].append(7)
    graph[8].append(16)
    graph[8].append(17)
    graph[16].append(17)
    graph[18].append(6)
    
 
    # add cost to each edge
    cost[(0, 4)] = 3
    cost[(0, 5)] = 9
    cost[(0, 16)] = 1
    cost[(2, 1)] = 2
    cost[(3, 1)] = 2
    cost[(4, 2)] = 1
    cost[(4, 3)] = 8
    cost[(4, 5)] = 2
    cost[(5, 8)] = 3
    cost[(5, 18)] = 2
    cost[(6, 3)] = 3
    cost[(6, 7)] = 2
    cost[(8, 16)] = 4
    cost[(8, 17)] = 4
    cost[(16, 17)] = 15
    cost[(18, 6)] = 1
    
    # set start state 
    start = 0
    
    # set goal state, there can be multiple goal states
    goal = [7, 18]
    
    # call uniform_search_cost function to get the minimum cost to reach gaol. Bonus points for path
    # ****** You have to implement this function *****
    min_cost_info = uniform_cost_search(start, goal, graph, cost)

    for node, info in min_cost_info.items():
        print(f'Minimum cost from {start} to {node} is {info["cost"]}')
      
        #   **** Bonus ****
        print(f'Path: {info["path"]}')
```
